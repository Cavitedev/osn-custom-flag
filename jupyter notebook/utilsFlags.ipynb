```python
import json
import requests
import os
```

```python
flagsPath = "./flags.json"

with open(flagsPath, "r", encoding="utf-8") as json_file:
    flags = json.load(json_file)
```

```python

countryCodes = flags.keys()
for countryCode in countryCodes:
    country = flags[countryCode]
    if(not 'regions' in country): continue
    for flagCode in country["regions"]:
        flagData = country["regions"][flagCode]
        if(not 'flag' in flagData): continue
        flag: str = flagData["flag"]
        if "static.wikia.nocookie" in flag:
            fileExtension = os.path.splitext(flag)[1] or ".png"
            response = requests.get(flag)
            os.makedirs(f"../{countryCode}", exist_ok=True)
            filename = f"{countryCode}/{flagCode}{fileExtension}"
            with open(f"../{filename}", "wb") as f:
                f.write(response.content)
                flagData["flag"] = f"https://raw.githubusercontent.com/Cavitedev/osn-custom-flag/main/{filename}"
                print(f"Downloaded {filename}")
                # flagData["flag"] = "base + country + '/' + flagCode + extension"
```

```python
json_data = json.dumps(
    flags, indent=4, ensure_ascii=False).encode('utf8').decode()

with open(flagsPath, "w") as json_file:
    json_file.write(json_data)

print("JSON data has been saved to", flagsPath)
```
